# Διασφάλιση Ποιότητας Αποτελεσμάτων Δοκιμών (Quality Assurance of Testing Results)


## Εισαγωγή

Ένα Εργαστήριο θα πρέπει να καθιερώσει και να λειρουργεί ως τμήμα του Συστήματος Ποιότητας του, να παρακολουθεί και να αξιολογεί την αναλυτική ικανόητα του αολι 'μέρα σε 'μερα (*day to day*) και από σειρά δειγμάτων (*batch to batch*) κατάλληλου επιπέδου σχήματα **Εσωτερικού Ελέγχου Ποιότητας (*Internal QC*)**. Πρέπει επίσης να συμμετέχει, όπου είναι δυνατόν σε κατάλληλα προγράμματα **Δοκιμών Ικανότητας (*Proficiency Testing - External QC)***. 

Σκοπός του ελέγχου ποιότητας είναι η διασφάλιση της σταθερότητας των αποτελεσμάτων από μέρα σε μέρα και η συμμόρφωση τους με προκαθορισμένα κριτήρια. 

Το πρότυπο ISO/IEC 17025:2015 § 7.7 απαιτεί εμέσως την εφαρμωγή Διαγραμμάτων ελέγχου ποιότητας ως εξής:

> *The laboratory shall have **quality control procedures** for monitoring the validity of tests and calibrations undertaken.*

> *The resulting data shall be **recorded** in such a way that **trends are detectable** and, where practicable, **statistical techniques shall be applied** to the reviewing of the results.*

> *The monitoring shall be planned and reviewed and may include, but is not limited to:*
> * *regular **use of certified reference materials** and/or **internal quality control**, using **secondary reference materials***
> * *participation in **interlaboratory comparison** or **proficiency-testing programmes***
> * ***replicate** tests or calibrations using the same, or different methods*
> * ***retesting** or recalibration of **retained items***
> *correlation of results for different characteristics of an item*

Τα **διαγράμματα ελέγχου ποιότητας** *(Quality Control Charts)* είναι ένα ισχυρό και εύχρηστο εργαλείο στον έλεγχο αναλύσεων ρουτίνας. Αυτά τα διαγράμματα εισήχθησαν πρώτα στις αυτοκινητοβιομηχανίες από τον Shewhart το 1931. Είτε για ξαφνικές μεταβολές είτε για αργή και σταδιακή πτώση του επιπέδου ποιότητας, τα διαγράμματα αυτά επιτρέπου την έγκαιρη διάγνωση και ταυτοποίηση τέτοιων προβλημάτων και επιτρέπουν άμεσες επεμβάσεις, μειώνοντας το ρίσκο ελαττωματικών προϊόντων και παραπόνων από πελάτες.

Η γενική αρχή τους είναι:

1. Λαμβάνουμε δείγματα ελέγχου κατά τη διάρκεια της παραγωγικής διαδικασίας
2. Μετράμε κάποιον δείκτη ποιότητας
3. Καταγράφουμε αυτές τις μετρήσεις σε διάγραμμα με προσημειωμένά ***όρια δράσης (action limits)*** και ***όρια προειδοποίησης (warning limits)***.

Στην Αναλυτική Χημεία ο δείκτης ποιότητας είναι μια τιμή στόχος. Αυτή μπορεί να είναι η πιστοποιημένη τιμή ενός πιστοποιημένου ή μη **υλικού αναφοράς** ή εάν δεν υπάρχουν διαθέσιμα υλικά αναφοράς, η μέση τιμή συχνά επαναλαμβανόμενων μετρήσεων δείγματος ελέγχου ποιότητας.

Γενικά θεωρούμε οτι τα δεδομένα τέτοιων μετρήσεων είναι κανονικά κατανεμημένα, και συνεπώς το $95,5\%$ των (ορθών ή αποδεκτών) τιμών βρίσκονται στο διάστημα $\mu\pm\;2\sigma$ και το $99,7\%$ των "ορθών" τιμών θα βρίσκονται εντός του διαστήματος $\mu\pm\;3\sigma$. Επομένω;ορίζουμε τα $\mu\pm\;2s$ ως τα **Όρια Προειδοποίησης (warning limits)** και τα $\mu\pm\;2s$ ως **Όρια Δράσης (action limits)**.

Υπάρχει μόλις $0,3\%$ πιθανότητα μία ορθή τιμή να βρεθεί εκτός ορίων δράσης, κάτι που ισοδυναμεί με 3 στις 1000 μετρήσεις. Επομένως η διαδικασία θα πρέπει γενικά να σταματήσει και να διερευνήσουμε για πιθανά σφάλματα όταν δούμε τέτοια απόκλιση.

Υπάρχει $4.5\%$ πιθανότητα μια ορθή τιμή να βρεθεί εκτός των ορίων προειδοποίησης. Αυτό δεν είναι κάτι υπερβολικά απίθανο, επομένως υπέρβαση αυτών των ορίων είναι μια προειδοποίηση και δεν χρειάζεται να κάνουμε κάτι άμεσα.

## Τυπική Απόκλιση

Μετρήσεις τυπικής απόκλισης θα πρέπει να είναι μεταξύ παρτίδων και επομένως κάθε σημείο σε τέτοιο διάγραμμα θα πρέπει επίσης να είναι μεταξύ παρτίδων. Στη πράξη, για να κατασκευάσουμε διάγραμμα ελέγχου ποιότητας θα πρέπει να κάνουμε μια προκαρτακτική εκτίμηση για ένα διάστημα περίππου 20 ημερών (ώστε να οριστούν η μέση τιμή και τα όρια). Δεν μπορούμε να χρησιμοποιήσουμε δεδομένα τυπική απόκλισης από πειράματα επαναληψημότητας (γιατί τα όρια τους είναι στη πράξη υπερβολικά μικρά), και ούτε από δεδομένα ενδιάμεση πιστότητας (ή αναπαραγωγημότητας) γιατί αυτά είναι υπερβολικά μεγάλα. Επομένως τα δεδομένα αυτά πρέπει να προκύψουν από μία προκαρτακτική επεξεργασία 20 ημερών.

Τα όρια δράσης και προειδοποίησης θα πρέπει να είναι κατάλληλα για το σκοπό της μεθόδου και δεν πρέπει να χρησιμοποιούνται "τυφλά" αλλά κατά περίπτωση μπορεί να προσαρμώζονται ώστε να είναι κατάλληλα για το σκοπό της μεθόδου.

### Διάγραμμα-Χ (X-Chart)

Λέγεται επίσης Διάγραμα Ελέγχου-Χ, Διάγραμα Ελέγχου Μέσης Τιμής,ή Διάγραμα Ελέγχου Μέσου. Είναι το πρώτο που χρησιμοποιήθηκε από τον Shewhart και συνήθως ελέγχει την ορθότητα και την ακρίβεια μέσο ανάλυσης συνθετικών δειγμάτων με γνωστή συγκέντρωση (πιστοποιημένου ή μη) υλικού αναφοράς. Ενακτικά μπορούμε να χρησιμοποιήσουμε και παραμέτρους της καμπύλης αναφοράς (σταθερός όρος ή συντελεστής συγκέντρωσης) για να ελένξουμε την σταθερότητας της βαθμονόμησης.

> **Στο Εργαστήριο στο οποίο εργάζεσθε  εφαρμόζατε Εσωτερικό Έλεγχο Ποιότητας, αλλά δεν δημιουργήσατε διαγράμματα ελέγχου (control chart). Ενόψει της προετοιμασίας του Εργαστηρίου σας  είναι ανάγκη να κατασκευάσετε αυτά τα διαγράμματα από τα ήδη υπάρχοντα δεδομένα.
Δεδομένα εσωτερικού ελέγχου ποιότητας με τη μεθοδολογία του δείγματος ελέγχου.
Ένα άγνωστο δείγμα, ικανοποιητικής ποσότητας και σταθερότητας, ομογενοποιήθηκε και αναλύθηκε σε 20 ανεξάρτητες αναλύσεις. Πάρθηκαν τα παρακάτω αποτελέσματα:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import markdown
import IPython.display as display
from IPython.display import HTML
import numpy as np
from collections import deque,namedtuple
from statistics import stdev, mean
from dataclasses import dataclass
from copy import deepcopy



# Styling
plt.style.use("fivethirtyeight")

CSS = """
.output {
    align-items: center;
}
"""

HTML('<style>{}</style>'.format(CSS))

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100

```

```python
df_A1 = pd.read_csv("Διαγράματα Ελέγχου Ποιότητας Τ.Α.1 Μέσου Όρου.csv",index_col = False, header=0)

# Display table as visualy pleasant Markdown
display.Markdown(df_A1.to_markdown())
```

```python
# Θα χρειαστούμε τα όρια, άρα την τυπική απόκλιση και επίσης τον μέσο όρο
aux_table = df_A1["Result"]
m = aux_table.mean()
sd = aux_table.std()

# Plot the base graph
## Warning Limits
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.axhline(y = m+2*sd,color="#f57b25",linestyle ='dashed', linewidth='1.5')
ax1.axhline(y = m-2*sd,color = "#f57b25",linestyle='dashed', linewidth='1.5')
ax2.axhline(y = m+2*sd,color="#f57b25",linestyle ='dashed', linewidth='1.5')
ax2.axhline(y = m-2*sd,color = "#f57b25",linestyle='dashed', linewidth='1.5')
## Action Limits
ax1.axhline(y=m+3*sd,color = 'r',linestyle='solid', linewidth='1.5')
ax1.axhline(y=m-3*sd,color='r',linestyle='solid',linewidth="1.5")
ax2.axhline(y=m+3*sd,color = 'r',linestyle='solid', linewidth='1.5')
ax2.axhline(y=m-3*sd,color='r',linestyle='solid',linewidth="1.5")
# Target Value
ax1.axhline(y=m,color='b', linestyle='solid',linewidth='1.5')
ax1.set_xlabel("Αριθμός Μέτρησης")
ax1.set_ylabel("Μέτρηση")
ax2.axhline(y=m,color='b', linestyle='solid',linewidth='1.5')
ax2.set_xlabel("Αριθμός Μέτρησης")
ax2.set_ylabel("Μέτρηση")
ax2.plot(df_A1["# Repeat Measurement"],df_A1["Result"], 'bo' )
ax1.set_title("Διάγραμα Ελέγχου Μέσου Όρου - Μόνο Όρια",y=1.08)
ax2.set_title("Διάγραμα Ελεγχου Μέσου Όρου - Τιμές",y=1.08)
ax2.set_ylim([75,95])
ax2.set_xlim([0,25])
ax1.set_xlim([0,25])
ax1.set_ylim([75,95])
plt.tight_layout()
plt.subplots_adjust(right=7,left=5,top=1.01)
# Add annotations to lines
plt.show()
```

Το παραπάνω δείγμα ελέγχου (control sample) χρησιμοποιήθηκε τις επόμενες 30 εργαστηριακές ημέρες για τη διασφάλιση των αποτελεσμάτων της μεθόδου και πάρθηκαν τα παρακάτω αποτελέσματα:

```python
df_A2 = pd.read_csv("Διαγράματα Ελέγχου Ποιότητας Τ.Α.2 Εκτός Ελέγχου.csv",index_col=False,header=0)
display.Markdown(df_A2.to_markdown())
```

Να τοποθετηθούν τα 30 σημεία επί του χάρτη ελέγχου και να σχολιασθούν τα σημεία (αν υπάρχουν) για τα οποία το Εργαστήριο έπρεπε να αναλάβει διερευνητικές ενέργειες. Την 16η και 26η ημέρα ο αναλυτής παρασκεύασε καινούργιο πρότυπο βαθμονόμησης. Πώς κρίνετε τη σταθερότητα του προτύπου αυτού διαλύματος;

```python
# Find outliers

out_of_bounds = df_A2[(df_A2["Result"]>=m+3*sd) | (df_A2["Result"]<=m-3*sd)]
out_of_warning = df_A2[((df_A2["Result"]>=m+2*sd) &(df_A2["Result"]<m+3*sd)) | ((df_A2["Result"]<=m-2*sd)&(df_A2["Result"]<m-2*sd))]
clear = df_A2[~(df_A2["Result"]>=m+3*sd) | (df_A2["Result"]<=m-3*sd)]
refreshed = pd.concat([df_A2[df_A2["A/A"] == 16], df_A2[df_A2["A/A"] == 26]],axis=0)
plt.plot(df_A2["A/A"],df_A2["Result"],'b-' ,linewidth=1.5)
plt.plot(clear["A/A"],clear["Result"],'bo',markersize=5, label = "Συμμορφούμενες")
plt.plot(out_of_bounds["A/A"],out_of_bounds["Result"],"o",markersize=5,markerfacecolor = "#f0100c", markeredgecolor="#f0100c",label = "Εκτός Ορίων")
plt.plot(out_of_warning["A/A"],out_of_warning["Result"] ,"o",markersize=5,markerfacecolor = "#f07a0c", markeredgecolor="#f07a0c", label = "Εκτός Ορίων Προειδοποίησης")
plt.axhline(y = m+2*sd,color="#f57b25",linestyle ='dashed', linewidth='1.5')
plt.axhline(y = m-2*sd,color="#f57b25",linestyle ='dashed', linewidth='1.5')
plt.axhline(y=m+3*sd,color = 'r',linestyle='solid', linewidth='1.5')
plt.axhline(y=m-3*sd,color='r',linestyle='solid',linewidth="1.5")
plt.plot(refreshed["A/A"],refreshed["Result"],"bv",label="Φρέσκα Δείγματα Ποιότητας",markersize=9)
plt.title("Διάγραμα Ελέγχου Ποιότητας - Τύπου Χ",y=1.08)
plt.xlim(0,35)
plt.ylim(75,100)
plt.xlabel("Αριθμός Μέτρησης")
plt.ylabel("Μέτρηση")
plt.xticks(ticks = np.arange(0,35,step=2) )
# plt.yticks(ticks = np.arange(75,100,step=2) )
plt.legend(loc="lower left",bbox_to_anchor = (1.08,0.3))
plt.show()
```

### Διάγραμμα Τιμής Λευκού (Blank Value Chart)

Είναι μια εξιδικευμένη εκδοχή του διαγράμματος μέσου. Αναλυουμε δείγμα που υποτίθεται οτι στερείται αναλύτη (το λευκό), οπότε παίρνουμε πληροφορίες για πιθανή επιμόλυνση αντιδραστηρίων, τη κατάσταση του αναλυτικού συστήματος και πιθανή επιμόλυνση από το περιβάλλον (ιδιαίτερα για εργαστήρια μοριακή βιολογίας). Εισάγουμε εδώ άμεσες μετρήσεις σημάτων όχι προβλεπόμενες από καμπύλη συγκεντρώσεις (αφού υποτίθεται οτι οι συγκετρώσεις θα πρέπει να είναι προσεγγιστικά μηδέν).

### Διάγραμα Ανάκτησης (Recovery Chart)

Εξαιτάζει την επιροή του υλικού μήτρας. Αναλύουμε πραγματικά δείγματα (ως έχουν), ενισχύουμε τα δείγματα με γνωστή συγκέντρωση αναλύτη ($\Delta C$) και επαναναλύουμε. Υπολογίζουμε την ανάκτηση βάση τύπου:
$$
\%Rec = \dfrac{C_{spiked}-C_{unspiked}}{\Delta C_{expected}}*100\%
$$

Αυτή η τιμή ανάκτησης σημειώνεται σε διάγραμμα. Τα διαγράματα αυτά ανιχνεύουν μόνο αναλογικά συστηματικά σφάλματα ενώ συστηματικά σταθερά σφάλματα δεν φαίνονται. Η τιμή στόχος είναι (κατά προσέγγιση) το $100\%$ και πρέπει να λάβουμε υπ' όψιν οτι ο ενισχυμένος αναλύτης μπορεί να δεσμεύεται δαιφορετικά στη μήτρα δείγματος.


### Διάγραμα Διαφοράς (Difference Chart)

Γίνεται ανάλυση δείγματος στην αρχή σειράς μετρήσεων και μετά στο τέλος. Η **αλγεβρική** (δηλαδή με πρόσημο) διαφορά των δύο μετρήσεων καταγράφεται στο διάγραμμα ελέγχου. Προφανώς η τιμή-στόχος είναι κατά προσέγγιση το μηδέν ενώ αν είναι σημαντικά διαφορετική από αυτό, τότε υποδηλώνεται ***ολίσθηση (drift)*** των αναλύσεων, ενός της σειράς. Τα διαγράμματα αυτά είναι κατάλληλα για έλεγχο της επαναληψημότητας και πιθανής ολίσθησης.

### Διάγραμμα CUSUM

Είναι το περιπλοκότερο διάγραμα ελέγχου ποιότητας, και βασίζεται στο *άθροισμα των διαφορών* από τιμή στόχο (**CU**SUM = **cu**mulative **sum**). Η τιμή στόχος αφαιρείται από κάθε ανάλυση και η προκύπτουσα διαφορά προστίθεται στο άθροισμα όλων των προηγούμενων διαφορών. Μια γενική εικόνα είναι ως εξής:

**Τ=80** **s=2.5**

|#N|c|c-T Cusum|
---|---|---
1|82|+2
2|79|-1|+1
3|80|0|+1
4|78|-2|-1

Εδώ τα όρια δεν είναι πλέον δύο γραμμές αλλά προκύπτουν ως μια νοερή "μασκα" σε σχήμα "ν" συνεστραμένου $90^ο$. Η "μασκα" αυτή ορίζεται από δύο τιμές, μορφής $Μ(d,\theta)$ όπου η $d$ παράμετρος ορίζει τη θέση της ("πάνω-κάτω") και η γωνία $\theta$ γωνία της (από τη κάθετη $d$ δηλαδή το μισό της συνολικής - δηλαδή ορίζει το πόσο "ανοιχτή" ή "κλειστή" είναι). Οι παράμετροι $d,\theta$ επιλέγονται έτσι ώστε ελάχιστοι "ψευδείς συναγερμοί" να γίνονται όταν η διαδικασία είναι υπό έλεγχο αλλά με μια σημαντική αλλαγή στο μέσο, γρήγορα αυτή εντοπίζεται.

Παρά τη σχετική περιπλοκότητα του, πλεονεκτεί καθώς δείχνει σε ποιό σημείο ακριβώς η διαδικασία βγήκε εκτός ελέγχου, γιατί η ανίχνευση γίνεται συντομότερα (μικρότερο απαιτούμενος αριθμός σημείων, μέχρι να εμφανιστεί απόκλιση) και γιατί η έκταση της πιθανής αλλαγής μπορεί να εκτιμηθεί από το μέσο, από τη κλίση του μέσου.

### Διαγράματα Ελέγχου Στόχου (Target Control Charts)

Εν αντιθέση με τα "κλασσικά" διαγράματα ποιότητας, τύπου Shewhart,τα διαγράματα ελέγχου ποιότητας στόχου, λειτουργούν με προκαθορισμένα κριτήρια ποιότητας και χωρίς στατικώς καθορισμένες τιμές. Τα όρια δίνονται από εξωτερικά καθορισμένα και ανεξάρτητα κριτήρια ποιότητας, σύμφωνα με την κατάλληλότητα για τον σκοπό για τον οποίο προορίζεται. Κάθε προηγούμενος τύπος διαγράματος μπορεί να μετατραπεί σε ελέγχου στόχου, μεταβάλλοντας τα όρια. Διαγράματα ελέγχου-στόχου είναι κατάλληλα αν:

* Δεν υπάρχει κανονική κατανομή από τις τιμές δειγμάτων ελέγχου (λόγο επίμονων καταστάσεων εκτός ελέγχου)
* Δεν υπάρχουν επαρκεί δεδομένα για στατιστικές αναλύσεις (σπανίως αναλυόμενες παράμετροι)
* Υπάρχουν εξωτερικά, καθορισμένα όρια τα οποία θα πρέπει να εφαρμωστούν, ώστε να διασφαλιστεί η ποιότητας των αναλυτικών αποτελεσμάτων
* Τα δείγματα ελέγχου είναι και τα όμοια και για τα δύο διαγράματα

Τα δε όρια μπορούν να προκύπτουν από:

* Νομικές απαιτήσεις
* Πρότυπα αναλυτικών μεθόδων και απαιτήσεις για τον εσωτερικό έλεγχο ποιότητας
* Η ελάχιστη ακρίβεια-ειδικά για το εργαστήριο- και η ορθότητα για την αναλυτική τιμή, οι οποίες θα πρέπει να διασφαλιστούν
* Από αξιολόγηση εσωτερικών (του εργαστηρίουν) δεδομένων, για τον ίδιο τύπο δείγματος

Κατασκευάζονται παρόμοια με τα κλασσικά, με ανώτερα και κατώτερα όρια, και η διαδικασία είναι εκτός ελέγχου μόνο όταν αυτά τα όρια υπερβαίνονται. Παρόλα αυτά όταν εμφανίζονται τάσεις στα δεδομένα, αυτές θα πρέπει να διερευνόνται και θα πρέπει να γίνονται ενέργειες για την καταπολέμηση τους.

## Διαγράματα Κινούμενου Μέσου Όρου
Είναι μια ποιό "δυναμική" εκδοχή του X-Chart όπου επαναϋπολογίζονται οι παράμετροι για κάθε νέα εισερχόμενη τιμή. Συνοπτικά η διαδικασία έχει ως εξής:

1. Εκτελούμε Ν διαδοχικές μετρήσεις, λαμβάνοντας τις κλασσικές παραμέτρους $(\hat{\mu},\hat{\sigma})$ - δηλαδή μέσο όρο και τυπική απόκλιση
2. Για κάθε νέα μέτρηση, απορρίπτουμε την παλαιότερη μέτρηση και εισάγεται η νέα
3. Επαναϋπολογίζουμε τις παραμέτρους  $(\hat{\mu},\hat{\sigma})$

```python
@dataclass(init=True, repr=True, eq=False, order=False, unsafe_hash=False, frozen=False)
class MovingMean:
    
    Vector = list[float]
    
    def __init__(self, start_pts : Vector, new_pts : Vector)->None:
        self.active_pts = deque(start_pts)
        self._cnt = len(self.active_pts)
        self.new_pts = deque(new_pts)
        self.new_pts_cp = deepcopy(self.new_pts)
        self._upper_bound = len(start_pts) + len(start_pts)
        self.lower_bound = len(start_pts)
        self.past_pts = deque()
        self.params = {"mu":[], "sd":[]}
        self._line_data = pd.DataFrame(data = {
            'upper_action':[],
            'lower_action':[],
            'upper_warning':[],
            'lower_warning':[],
            'average':[],
            'idx':[]
        })
        self.__init_table()
        
        
    def __init_table(self)->None:
        self._line_data.columns = ["upper_action","lower_action","upper_warning","lower_warning","average","idx"]


    def __reshape_table(self)->None:
        self._line_data = self._line_data.astype({"idx":"int32"})
        self._line_data = self._line_data.set_index('idx')

    def __set_table(self,graph_vars : namedtuple)->None:
        new_vars = {
            "upper_action":graph_vars.mu+3*graph_vars.sd,
            "lower_action":graph_vars.mu-3*graph_vars.sd,
            "upper_warning": graph_vars.mu+2*graph_vars.sd,
            "lower_warning": graph_vars.mu-2*graph_vars.sd,
            "average":graph_vars.mu,
            "idx":self._cnt
        }
        self._line_data = self._line_data.append(new_vars,ignore_index=True)
        
    def __bitshift(self)->None:
        self.active_pts.append(self.new_pts.pop())
        self.past_pts.append(self.active_pts.popleft())
    
    def __get_params(self)->namedtuple:
        vars = namedtuple("Vars", ['mu','sd'])
        c = self.active_pts
        m = mean(c)
        sd = stdev(c)
        self.params["mu"].append(m)
        self.params["sd"].append(sd)      
        vars.mu = m
        vars.sd = sd
        return vars



    def __render__(self)->None:
        fig,ax = plt.subplots()
        ax.set_xlim(self.lower_bound,self._upper_bound)
        for idx,col in self._line_data.iterrows():
            if idx != 20:
                ax.plot((idx-1,idx),(self._line_data.loc[idx]["upper_action"], self._line_data.loc[idx]["upper_action"]),"r--", linewidth=1)
                ax.plot((idx-1,idx-1),(self._line_data.loc[idx-1]["upper_action"],self._line_data.loc[idx]["upper_action"]),"r--", linewidth=1)
                ax.plot((idx-1,idx),(self._line_data.loc[idx]["lower_action"], self._line_data.loc[idx]["lower_action"]),"r--", linewidth=1)
                ax.plot((idx-1,idx-1),(self._line_data.loc[idx-1]["lower_action"],self._line_data.loc[idx]["lower_action"]),"r--", linewidth=1)
                
                ax.plot((idx-1,idx),(self._line_data.loc[idx]["upper_warning"], self._line_data.loc[idx]["upper_warning"]),"y--", linewidth=1)
                ax.plot((idx-1,idx-1),(self._line_data.loc[idx-1]["upper_warning"],self._line_data.loc[idx]["upper_warning"]),"y--", linewidth=1)
                ax.plot((idx-1,idx),(self._line_data.loc[idx]["lower_warning"], self._line_data.loc[idx]["lower_warning"]),"y--", linewidth=1)
                ax.plot((idx-1,idx-1),(self._line_data.loc[idx-1]["lower_warning"],self._line_data.loc[idx]["lower_warning"]),"y--", linewidth=1)

                ax.plot((idx-1,idx),(self._line_data.loc[idx]["average"],self._line_data.loc[idx]["average"]), "b-", linewidth=1)
                ax.plot((idx-1,idx-1),(self._line_data.loc[idx-1]["average"], self._line_data.loc[idx]["average"]), "b-", linewidth=1)

        
        ax.set_xlabel("Αριθμός Μέτρησης")
        ax.set_ylabel("Μέτρηση")
        ax.set_xticks(np.arange(20,55,step = 5))
        ax.set_title("Διάγραμμα QC Κινούμενου Μέσου Όρου")
        ax.plot(np.arange(20,len(self.new_pts_cp) + 20,step=1), (self.new_pts_cp), "kP")
        plt.show()
                
            
            
            
    def __call__(self)->None:
        graph_vars = self.__get_params()
        self.__set_table(graph_vars)
        new_pts = deepcopy(self.new_pts)
        for elem in new_pts:
            self._cnt += 1
            self.__bitshift()
            self.__set_table(self.__get_params())
        self.__reshape_table()
        self.__render__()

x = pd.read_csv("Διαγράματα Ελέγχου Ποιότητας Τ.Α.1 Μέσου Όρου.csv",index_col=False)
y = pd.read_csv("Διαγράματα Ελέγχου Ποιότητας Τ.Α.2 Εκτός Ελέγχου.csv",index_col=False)


x = MovingMean(list(x["Result"]),list(y["Result"]))
x()
```

### Διπλά Διαγράματα (Twin Plot)

Μια άλλη προσέγγιση σε Κλινικά, κυρίως εργαστήρια, είναι το **διάγραμα Youden** (σπανιότερα twin plot ή A+B plot ή  δύο δειγμάτων ή συνεργαστηκός έλεγχος δύο δειγμάτων). Η περίπτωση διαγράμματο διασπωράς (scatter plot) που χρησιμοποιείται κυρίως για τη σύγκριση αποτελεσμάτων διαφορετικών (κατά ζεύγοι) εργαστηρίων στο ίδιο διάγραμμα. Μπορεί να δώσει πληρωφορίες για συστηματικά σφάλματα στις μετρήσεις, και να απαντήσει σε ερωτήματα όπως

* Είναι τα εργαστήρια ισοδύναμα?
* Ποιά εργαστήρια δίνουν έκτροπες τιμές (εφόσον υπάρχουν)?
* Οι ασυμφωνιες των δεδομένων προέρχονται προτίστως μεταξύ των εργαστηρίων (προβήματα αναπαραγωγημότητας) ή εντός εργαστηρίου (προβλήματα επαναληψημότητας) ?

Υπάρχουν δύο προσεγγίσεις για αυτούς τους ελέγχους (κυρίως). Στη μία περίπτωση το ίδιο υλικό στέλνεται σε όλα τα εμπλεκόμενα εργαστήρια και αυτά θα εκτελέσουν ένα προσδιορισμό το καθένα. Εναλλακτικά στέλνονται δύο υλικά σε κάθε εργαστήριο. Τότε κατασκευάζουμε απλό διάγραμμα διασποράς τυπου ΧΥ με το αποτέλεσμα του ενός εργαστηρίου στον ένα άξονα και του άλλου στον δεύτερο άξονα (αν στάλθηκαν δύο υλικά τότε το αποτέλεσμα για κάθε υλικό μπαίνει σε έναν άξονα). Τότε κάθε σημείο θα είναι ένα ζεύγος αποτελεσμάτων διαφορετικών εργαστηρίων (για δύο υλικά είναι ζεύγος μετρήσεων κάθε εργαστηρίου). Τυπικά χαράσσεται και μια γραμμή αναφοράς $45^ο$ που περνά από την αρχή των αξόνων. Θα πρέπει επίσης να γνωρίζουμε και τις τυπικές αποκλίσεις αποτελεσμάτων διεργαστηριακών συγκρίσεων (έστω $\hat{\sigma_{_1}},\;\;\hat{\sigma_{_2}}$). Τα αρχικά διαγράμματα Youden εμπεριείχαν απλά τις τιμές στου άξονες (δηλαδή μορφής $(x_{1_{i}},\;x_{2_{i}})$), οστόσο το ISO 13528:2005 ορίζει ένα διάγραμμα Youden ως ένα διάγραμμα διασποράς $(z_{1i},\;z_{2i})$, όπου τα z-scores προκύπτουν από διεργαστηριακές συγκρίσεις:


$$
z_{1i} = \dfrac{x_{1i}-X_1}{\hat{\sigma_{_1}}},\;\;\;z_{2i} = \dfrac{x_{2i}-X_2}{\hat{\sigma_{_2}}}
$$

Τα ($\hat{\sigma_{_1}},\;\;\hat{\sigma_{_2}}$) είναι οι εκτιμήτριες των τυπικών αποκλίσεων πληθυσμών, όπως αυτές προέκυψαν από τα δεδομένα διεργαστηριακών συγκρίσεων.

> **Παρατήρηση:** \
> Κατά ISO 13528:2005 η μεθοδολογία δεν είναι ανθεκτική (robust) σε έκτροπες τιμές και γενικότερ η προσέγγιση με z-score έχει κανονικοποιημένες παραμέτρους, ανεξάρτητες μεγέθους των τιμών, αλλά είναι δυσκολότερη η παρατήρηση συνολικών τάσεων (για παράδειγμα οι έκτροπες εμφανίζουν μια τάση να συγκεντρώνονται γύρω από συγκεκτριμένα σημεία και δεν είναι άμεσα ξεκάθαρη η σχέση μεγέθους μετρούμενων με την διασπορά τους).

Σε ένα διάγραμμα Youden ένας κύκλος γύρο από την αρχή των αξόνων (ή μια έλειψη) ένα όριο διάκρισης σφαλμάτων. Κατά ISO 13528:2005 δεδομένων $n$ συμμετεχόντων, $\overline{x_{1_i}}$ και $\overline{x_{2_i}}$ τις μέσες τιμές για τα δύο δείγματα, $s_1$ και $s_2$ τις αντίστοιχες τυπικές αποκλίσεις, και $\hat{\rho}$ τον συντελεστή συσχέτησης δείγματος, τότε:

$$
\def\arraystretch{5}
\begin{array}{cc}
\overline{x_1}=\frac 1n\sum_{i=1}^nx_{1_i}&&\overline{x_2}=\frac 1n\sum_{i=1}^nx_{2_i}\\
s_1=\sqrt{\dfrac1{n-1}\sum_{i=1}^n{(x_{1_i}-\overline{x_{1}})^2}}&&s_2=\sqrt{\dfrac1{n-1}\sum_{i=1}^n{(x_{2_i}-\overline{x_{2}})^2}}\\
\hat{\rho} = \dfrac 1{s_1s_2}\dfrac 1{n-1} \sum_{i=1}^{n}{(x_1{_i}-\overline{x_1})(x_2{_i}-\overline{x_2})}
\end{array}
$$

Σε μια άλλη εκδοχή σε κάθε έλεγχο μετρόνται δύο δείγματα ελέγχου, ένα υψηλής και ένα χαμηλής τιμής και σε διάγραμμα σημειόνται τα σημεία με συντεταγμένες τις μετρήσεις καθώς και οι τιμές στόχοι και οι γραμμές ορίων $\pm 2SD$ ΚΑΙ $\pm 4SD$. Το σύστημα τότε θα βρίσκεται εντός ελέγχου όταν κάθε σημείο (δηλαδή ζεύγος μετρήσεων υψηλού και χαμηλού προτύπου), βρίσκεται εντός των ορίων $\pm 2SD$. Δηλαδή ελέγχου αν τα υψηλά και τα χαμηλά βρίσκονται ταυτόχρονα εντός των ορίων $\mu\pm 2SD$, με μ την τιμή-στόχο.

> Σε μία δοκιμή αποφασίσθηκε να χρησιμοποιούνται 2 δείγματα ελέγχου, διαφορετικής συγκέντρωσης. Για το σκοπό αυτό αγοράσθηκαν 2 πρότυπα αναφοράς με αναφερόμενες τιμές: Α) 45,2±2,3 και Β) 75,6±4,1.

## Τα Δείγματα Ελέγχου

Τα δείγματα ελέγχου είναι απαραίτητα για την κατασκευή διαγράματος ελέγχου ποιότητας. Θα πρέπει να πληρούν ορισμένες απαιτήσεις, δηλαδή θα πρέπει να:

* είναι κατάλληλα για παρακολούθηση σε μακρές χρονικές περιόδους
* είναι αντιπροσωπευτικές για τη μήτρα και τον αναλύτη (δεδομένων συγκεντρώσεων)
* έχουν συγκέντρωση εντός ορίων, όταν είναι εφικτό
* είναι επερκούς ποσότητας για επαρκή χρονική περίοδο
* είναι σταθερά για πολλούς μήνες
* μην εμφανίζουν απώλειες, λόγο περιέκτη
* μην εμφανίζουν αλλαγές, όταν λαμβάνονται υπο-δείγματα από αυτά

### Πρότυπα Διαλύματα

Μπορούμε να χρησιμοποιήσουμε πρότυπα διαλύματα για να επιβεβαιόσουμε την βαθμονόμηση και το δείγμα ελέγχου θα πρέπει να είναι εντελώς ανεξάρτητο από τα διαλύματα βαθμονόμησης (όχι από διαχωρισμό σε υποδείγματα πχ). Δεν θα πρέπει να ανιχνευεται επίδραση από το μητρικό υλικό. Δίνουν περιορισμένο έλεγχο ακρίβειας και ορθότητας.

### Λευκά Δείγματα

Πρόκειται για δείγματα, τα οποία πιθανότατα δεν περιέχουν τον αναλύτη. Μπορούν να βοηθήσουν στην ανίχνευση σφαλμάτων λόγο:

* αλλαγών σε αντιδραστήρια
* νέων παρτίδων αντιδραστηρίων
* μεταφοράς
* ολίσθησης παραμέτρων συστήματος

Οι τιμές λευκού στην αρχή και στο τέλος σειράς αναλύσεων μπορούν να βοηθήσουν στην ταυτοποίηση κάποιων συστηματικών τάσεων

### Πραγματικά Δείγματα

Αναλύοντια πολλαπλές φορές, σε περιπτώσεις διαγραμάτων εύρους και διαφορών. Ανάλογα με τη περίπτωση μπορεί να χρειαστούν δαιφορετικά διαγράματα για διαφορετικές μήτρες. Δεν επιτρέπουν τον έλεγχο της *ορθότητας* αλλά επιτρέπουν ταχύτητο έλεγχο της *ακρίβειας*.

### Εμβολιασμένα Δείγματα

Πρόκειται για πραγματικά δείγματα, που ενισχύονται με προκαθορισμένη ποσότητα αναλύτη. Είναι αποκλειστική επιλογή για διαγράματα ελέγχου ανάκτησης και βοηθούν στον εντοπισμό επιδράσεων λόγο μητρικού υλικού. Κατά περίπτωση, μπορεί να είναι απαραίτητη η κατασχευή χωριστών διαγραμάτων ανά μήτρα, ενώ η ουσία ενίσχυσης θα πρέπει να είναι αντιπροσωπευτική για τον αναλύτη στο δείγμα (για παράδειγμα ορισμένες ουσίες μπορεί να δεσμεύονται από ενδογεννείς πρωτεϊνες, οπότε ελέγχεται μόνο η ελέυθερη μορφή). Επιτρέπουν περιορισμένο έλεγχο της ορθότητας.

### Συνθετικά Δείγματα

Σπανιώς χρησιμοποιούνται, διότι τα μητρικά υλικά, συνήθως είναι υπερβολικά σύνθετα και μεταβλητα για να αναπαραχθούν αποτελεσματικά, συνθετικά. Στις σπάνιες περιπτώσεις που αποδειχθούν αντιπροσωπευτικά των πραγματικών, τότε επιτρέπουν τον έλεγχο τόσο της ορθότητας, όσο και της ακρίβειας.

### Υλικά Αναφοράς

Τα *πιστοποιημένα υλικά αναφοράς (certified reference materials, CRM's)* είναι τα πλέον ιδανικά ως δείγματα ελέγχου, αλλά συχνά είναι απαγορευτικού κόστους (τα δείγματα ελέγχου ποιότητας καταναλώνονται ταχύτητα σε εργαστήρια ρουτίνας) ή σε πολλές περιπτώσεις κλινικών εργαστηρίων, δεν είναι καν εμπορικά διαθέσιμα. Μια συχνά ικανοποιητική εναλλακτική είναι τα εσωτερικα (in-house) υλικά αναφοράς. Αυτά μπορούν να διατηρούν την αξιοπιστία τους μέσω συχνών συγκρίσεων με CRM και αν η τιμή τους είναι γνωστή, τότε δίνουν δεδομένα για έλεγχο ορθότητας. Ένα τέτοια παράδειγμα είναι υπολοιπωμενα υλικά από διεργαστηριακά σχήματα ελέγχου ικανότητας.

## Συμπεράσματα

Υπάρχουν πολλές επιλογές, και η τελική απόφαση είναι στη κρίση του διευθυντή. Γενικά όσο συχνότερη μια ανάλυση, τόσο περισσότερο νόημα έχει το διάγραμ ελέγχου ποιότητας και αν οι αναλύσεις γίνονται πάντα σε ίδια μήτρα, τότε θα πρέπει να συμπεριληφθεί και η προετοιμασία του δείγματος (πχ με εμβολιασμένα ή λευκά δείγματα). Αν η μήτρα ποικίλει, τότε το διάγραμα θα πρέπει να περιοριστεί στη μέτρηση μόνο.

Σε κάποιες περιπτώσεις, η καθημερινή βαθμονόμηση δίνει συντελεστές που μπορούν να ενσωματωθούν σε διαγράματα ελέγχου ποιότητας, με ελάχιστη προσπάθεια. Μερικές φορές η νομοθεσία απαιτεί κατασκευή τέτοιων διαγραμάτων ως ελάχιστη προϋπόθεση αξιοπιστίας, οπότε η επιπλέον προσπάθεια κατασκευής διαγράματος είναι ελάχιστη και μπορεί να αυξήσει σημαντικά την αξιοπιστία των μετρήσεων (άρα και του εργαστηρίου).

Τα δεδομένα αυτών των διαγραμάτων μπορούν να χρησιμοποιηθούν από το εργαστήριο ως απόδηξη των τεχνικών ικανοτήτων του σε τρίτους, όπως πιθανός πελάτες ή ελενκτές.
